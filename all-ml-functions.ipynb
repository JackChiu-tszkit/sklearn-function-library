1.Classification: 

```python
"KNeighborsClassifier（K 近邻）"
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"SGDClassifier（随机梯度下降分类器）"
from sklearn.linear_model import SGDClassifier

model = SGDClassifier(random_state=1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"Naive Bayes（朴素贝叶斯）"
from sklearn.naive_bayes import MultinomialNB  # 适用于文本数据

model = MultinomialNB()
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"SVC（支持向量机分类）"
from sklearn.svm import SVC

model = SVC(kernel='rbf', random_state=1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"Ensemble Classifiers（集成分类器，如随机森林）"

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

2. 回归 (Regression)

```python
"SGDRegressor（随机梯度下降回归）"
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(random_state=1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"Lasso（Lasso 回归）"
from sklearn.linear_model import Lasso

model = Lasso(alpha=0.1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)

```

```python
"ElasticNet（弹性网络回归）"
from sklearn.linear_model import ElasticNet

model = ElasticNet(alpha=0.1, l1_ratio=0.5)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"SVR（支持向量机回归）"
from sklearn.svm import SVR

model = SVR(kernel='rbf')
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"RidgeRegression（岭回归）"
from sklearn.linear_model import Ridge

model = Ridge(alpha=1.0)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

```python
"Ensemble Regressors（集成回归，如随机森林回归）"
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=1)
model.fit(train_x, train_y)
predictions = model.predict(valid_x)
```

3. 聚类 (Clustering)

```python
"KMeans（K 均值聚类）"
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, random_state=1)
model.fit(train_x)
predictions = model.predict(valid_x)
```

```python
"GMM（高斯混合模型）"
from sklearn.mixture import GaussianMixture

model = GaussianMixture(n_components=3, random_state=1)
model.fit(train_x)
predictions = model.predict(valid_x)
```

```python
"MiniBatch KMeans（小批量 K 均值）"
from sklearn.cluster import MiniBatchKMeans

model = MiniBatchKMeans(n_clusters=3, random_state=1)
model.fit(train_x)
predictions = model.predict(valid_x)
```

```python
"MeanShift（均值漂移）"
from sklearn.cluster import MeanShift

model = MeanShift()
model.fit(train_x)
predictions = model.predict(valid_x)
```

```python
"VBGMM（变分贝叶斯高斯混合模型）"
from sklearn.mixture import BayesianGaussianMixture

model = BayesianGaussianMixture(n_components=3, random_state=1)
model.fit(train_x)
predictions = model.predict(valid_x)
```

4. 降维 (Dimensionality Reduction)

```python
"Randomized PCA（随机 PCA）"
from sklearn.decomposition import PCA

model = PCA(n_components=2, svd_solver='randomized')
model.fit(train_x)
reduced_data = model.transform(valid_x)
```

```python
"Isomap（等度量映射）"
from sklearn.manifold import Isomap

model = Isomap(n_components=2)
model.fit(train_x)
reduced_data = model.transform(valid_x)
```

```python
"Spectral Embedding（谱嵌入）"
from sklearn.manifold import SpectralEmbedding

model = SpectralEmbedding(n_components=2, random_state=1)
model.fit(train_x)
reduced_data = model.transform(valid_x)
```

```python
"LLE（局部线性嵌入）"
from sklearn.manifold import LocallyLinearEmbedding

model = LocallyLinearEmbedding(n_components=2, method='standard')
model.fit(train_x)
reduced_data = model.transform(valid_x)
```
